```python
import numpy as np
print("NumPy loaded successfully!")
```

```python
def print_matrix(name, matrix):
    print(f"\n{name}:")
    print("-" * 30)
    for row in matrix:
        print("  " + "  ".join(f"{val:8.2f}" for val in row))
    print("-" * 30)
    print(f"Shape: {matrix.shape[0]} x {matrix.shape[1]}")

def input_matrix(name):
    print(f"\nEnter {name}:")
    rows = int(input("Number of rows    : "))
    cols = int(input("Number of columns : "))
    print("Enter values row by row (space separated):")
    data = []
    for i in range(rows):
        while True:
            try:
                row_input = input(f"Row {i+1} : ").strip().split()
                if len(row_input) != cols:
                    print(f"Enter exactly {cols} values!")
                    continue
                data.append(list(map(float, row_input)))
                break
            except ValueError:
                print("Numbers only please!")
    matrix = np.array(data)
    print_matrix(f"{name} entered", matrix)
    return matrix
```

```python
def addition():
    print("\n--- MATRIX ADDITION ---")
    A = input_matrix("Matrix A")
    B = input_matrix("Matrix B")
    if A.shape != B.shape:
        print(f"Error: Shapes {A.shape} and {B.shape} must be same for addition!")
        return
    result = A + B
    print_matrix("Matrix A", A)
    print_matrix("Matrix B", B)
    print_matrix("Result A + B", result)

def subtraction():
    print("\n--- MATRIX SUBTRACTION ---")
    A = input_matrix("Matrix A")
    B = input_matrix("Matrix B")
    if A.shape != B.shape:
        print(f"Error: Shapes {A.shape} and {B.shape} must be same for subtraction!")
        return
    result = A - B
    print_matrix("Matrix A", A)
    print_matrix("Matrix B", B)
    print_matrix("Result A - B", result)

def multiplication():
    print("\n--- MATRIX MULTIPLICATION ---")
    A = input_matrix("Matrix A")
    B = input_matrix("Matrix B")
    if A.shape[1] != B.shape[0]:
        print(f"Error: Columns of A ({A.shape[1]}) must equal Rows of B ({B.shape[0]})")
        return
    result = np.dot(A, B)
    print_matrix("Matrix A", A)
    print_matrix("Matrix B", B)
    print_matrix("Result A x B", result)

def transpose():
    print("\n--- MATRIX TRANSPOSE ---")
    A = input_matrix("Matrix A")
    result = A.T
    print_matrix("Original Matrix A", A)
    print_matrix("Transpose of A", result)
    print(f"Shape changed: {A.shape} to {result.shape}")

def determinant():
    print("\n--- MATRIX DETERMINANT ---")
    print("Note: Only works on square matrices (2x2, 3x3, etc.)")
    A = input_matrix("Matrix")
    if A.shape[0] != A.shape[1]:
        print(f"Error: Matrix must be square! Got {A.shape}")
        return
    det = np.linalg.det(A)
    print_matrix("Matrix A", A)
    print(f"\nDeterminant = {det:.4f}")
    if det == 0:
        print("Determinant is 0 - Matrix is Singular (no inverse)")
    else:
        print("Matrix is Non-Singular (inverse exists)")
```

```python
while True:
    print("\n========== MATRIX OPERATIONS TOOL ==========")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Transpose")
    print("5. Determinant")
    print("6. Exit")
    print("=============================================")

    choice = input("Enter choice (1-6): ").strip()

    if choice == '1':
        addition()
    elif choice == '2':
        subtraction()
    elif choice == '3':
        multiplication()
    elif choice == '4':
        transpose()
    elif choice == '5':
        determinant()
    elif choice == '6':
        print("Goodbye!")
        break
    else:
        print("Invalid choice! Enter 1 to 6.")
        continue

    again = input("\nGo back to menu? (yes/no): ").strip().lower()
    if again != 'yes':
        print("Goodbye!")
        break5
```
